```python
from tkinter import *
import sqlite3
from tkinter import messagebox
from tkinter import ttk

#-----------------------------STARTING OF PROJECT--------------------------------------------------------------
class SetUp(Tk):
    def __init__(self):
        Tk.__init__(self)
        self._frame = None
        self.session = None
        self.geometry('600x420')
        self.resizable(1,1)
        self.title("Airline System")
        self.SwitchFrame(WelcomePage)
        
    def SwitchFrame(self,FrameClass):
        '''Function to switch between the frames for different functionality '''
        NewFrame = FrameClass(self)
        if self._frame is not None:
            self._frame.destroy()
        self._frame = NewFrame
        self._frame.pack()

#-------------------------------FIRST PAGE OF PROJECT------------------------------------------------------------
class WelcomePage(Frame):
    def __init__(self, master):
        Frame.__init__(self, master)
        master.title("Welcome Page")
        Label(self, text = "***** Airline Booking System *****",font=('Helvetica', 18, "bold"),bg='white',fg='blue')\
        .pack(side="top",fill=X)
        Button(self,text='press here to continue',bg='black',fg='white',command=lambda: master.SwitchFrame(HomePage)).pack(fill=X)

#--------------------------------LOGIN OR REGISTRATION CHOICE PAGE---------------------------------------------------------
class HomePage(Frame):
    def __init__(self, master):
        Frame.__init__(self, master)
        master.title("Home Page")
        Label(self, text = "***** Home Page *****",font=('Helvetica', 18, "bold"),bg='white',fg='blue').grid(row=1)
        Button(self,text='Login',bg='black',fg='white',width=20,command=lambda: master.SwitchFrame(LogIn)).grid(row=2)
        Button(self,text='Register',bg='white',fg='black',width=20,command=lambda: master.SwitchFrame(Register)).grid(row=3)
        Button(self,text='Quit',bg='black',fg='white',width=20,command=master.destroy).grid(row=4)

#------------------------------LOG IN PAGE----------------------------------------------------------------------------
class LogIn(Frame):
    def __init__(self, master):
        Frame.__init__(self, master)
        master.title("Login Page")
        Label(self, text = "***** Login Page *****",font=('Helvetica', 18, "bold"),bg='white',fg='blue')\
        .grid(row=0,columnspan=2)

        Label(self, text = "Username:", font=('arial',15), bd=15).grid(row=1,sticky="w")
        Label(self, text = "Password:", font=('arial', 15), bd=15).grid(row=2,sticky="w")

        self.usernameEntry = Entry(self, font=(3),bg='powder blue')
        self.usernameEntry.grid(row=1, column=1)
        self.passwordEntry = Entry(self, show="*", font=(3),bg='powder blue')
        self.passwordEntry.grid(row=2, column=1)

        Button(self,text='Login',bg='black',fg='white',width=15,command=self.verifyInfo).grid(row=3, column=0)
        Button(self,text='Cancel',bg='black',fg='white',width=15,command=lambda: master.SwitchFrame(HomePage)).grid(row=3, column=1)

    def verifyInfo(self):
        '''verify users login information'''
        try:
            mydb=sqlite3.connect("AirlineDatabase.db")
            cur=mydb.cursor()
            temp=(self.usernameEntry.get(),self.passwordEntry.get())
            info=cur.execute("select * from loginInfo where userId=? and password=?",temp)
            if len(list(info))==1:
                self.master.session=self.usernameEntry.get()
                messagebox.showinfo("Login Successfull","Login Successfull")
                self.master.SwitchFrame(UserWelcomePage)
            else:
                messagebox.showerror("Invalid Info","Invalid id or password")
            mydb.commit()
            mydb.close()
        except:
            messagebox.showerror("Invalid Info","SQL ERROR")

#-------------------------------REGISTRATION PAGE--------------------------------------------------------------------
class Register(Frame):
    def __init__(self, master):
        Frame.__init__(self, master)
        master.title("Registration Page")

        Label(self, text = "***** Customer Registration Page *****",font=('Helvetica', 18, "bold"),bg='white',fg='blue')\
        .grid(row=0,columnspan=2)
        Label(self, text = "User Id :", font=('arial',15), bd=15).grid(row=1,sticky="w")
        Label(self, text = "User Name :", font=('arial',15), bd=15).grid(row=2,sticky="w")
        Label(self, text = "Mobile :", font=('arial',15), bd=15).grid(row=3,sticky="w")
        Label(self, text = "Email Id :", font=('arial',15), bd=15).grid(row=4,sticky="w")
        Label(self, text = "Address :", font=('arial',15), bd=15).grid(row=5,sticky="w")
        Label(self, text = "Password :", font=('arial',15), bd=15).grid(row=7,sticky="w")

        self.userIdEntry = Entry(self, font=(3),bg='powder blue')
        self.userIdEntry.grid(row=1, column=1,sticky="w")
        self.userNameEntry = Entry(self, font=(3),bg='powder blue')
        self.userNameEntry.grid(row=2, column=1,sticky="w")
        self.userMobEntry = Entry(self, font=(3),bg='powder blue')
        self.userMobEntry.grid(row=3, column=1,sticky="w")
        self.userMailEntry = Entry(self, font=(3),bg='powder blue')
        self.userMailEntry.grid(row=4, column=1,sticky="w")
        self.userAddressEntry=Entry(self,font=(3),bg='powder blue')
        self.userAddressEntry.grid(row=5, column=1,sticky="w")
        self.userPassEntry = Entry(self, show="*", font=(3),bg='powder blue')
        self.userPassEntry.grid(row=7, column=1,sticky="w")

        Button(self,text='Register',bg='black',fg='white',width=15,command=self.doRegistration)\
        .grid(row=8, column=0,sticky='w')
        Button(self,text='Cancel',bg='black',fg='white',width=15,\
               command=lambda: master.SwitchFrame(HomePage)).grid(row=8, column=1,sticky='w')
    
    def doRegistration(self):
        '''store user information to database'''
        try:
            mydb=sqlite3.connect("AirlineDatabase.db")
            cur=mydb.cursor()
            tempUser=(self.userIdEntry.get(),self.userNameEntry.get(),self.userMobEntry.get(),self.userMailEntry.get(),self.userAddressEntry.get())
            tempLogin=(self.userIdEntry.get(),self.userPassEntry.get())
            cur.execute('insert into loginInfo values(?,?)',tempLogin)
            cur.execute('insert into userInfo values(?,?,?,?,?)',tempUser)
            mydb.commit()
            mydb.close()
            messagebox.showinfo("Registration status","Registration Successfull")
            self.master.SwitchFrame(LogIn)
        except:
            messagebox.showerror("Invalid Info","Provide Valid info")

#------------------------------------USER WELCOME PAGE--------------------------------------------------------------
class  UserWelcomePage(Frame):
    def __init__(self, master):
        Frame.__init__(self, master)
        master.title("User Welcome Page")

        Label(self, text = "***** Welcome To Airline *****",font=('Helvetica', 18, "bold"),bg='white',fg='blue').grid(row=0,columnspan=2)

        menu=Menu(master)
        master.config(menu=menu)

        bookingMenu=Menu(menu,tearoff=False)
        menu.add_cascade(label='Booking',menu=bookingMenu)
        bookingMenu.add_command(label='Add Booking',command=lambda: master.SwitchFrame(BookFlight))
        bookingMenu.add_command(label='Search Booking',command=lambda: master.SwitchFrame(SearchBooking))
        bookingMenu.add_command(label='Delete Booking',command=lambda: master.SwitchFrame(DeleteBooking))
        bookingMenu.add_separator()

        passengerInfoMenu=Menu(menu,tearoff=False)
        menu.add_cascade(label='PassengerInfo',menu=passengerInfoMenu)
        passengerInfoMenu.add_command(label='Details',command=lambda: master.SwitchFrame(UserDetail))
        passengerInfoMenu.add_command(label='update info',command=lambda: master.SwitchFrame(UpdateDetail))
        passengerInfoMenu.add_separator()
        
        Airline=Menu(menu,tearoff=False)
        menu.add_cascade(label='Airline',menu=Airline)
        Airline.add_command(label='Search Airline',command=lambda: master.SwitchFrame(SearchAirline))
        
        logoutMenu=Menu(menu,tearoff=False)
        menu.add_cascade(label='Logout',menu=logoutMenu)
        logoutMenu.add_command(label='Logout',command=self.doLogout)
        
        quitMenu=Menu(menu,tearoff=False)
        menu.add_cascade(label='Quit',menu=quitMenu)
        quitMenu.add_command(label='Quit',command=master.destroy)
    
    def doLogout(self):
        '''log out from the account'''
        msg=messagebox.askquestion("logout","Sure you want to logout ?")
        if msg=='yes':
            self.master.config(menu="")
            self.master.session=None
            self.master.SwitchFrame(HomePage)
        else:
            self.master.SwitchFrame(UserWelcomePage)
        
    def doNothing(self):
        pass

#------------------------BOOKING PAGE-----------------------------------------------------------------------------
class  BookFlight(Frame):
    def __init__(self, master):
        Frame.__init__(self, master)
        master.title("Book Flight")

        Label(self, text = "***** Book Your Flight *****",font=('Helvetica', 18, "bold"),bg='white',fg='blue').grid(row=0,columnspan=2,sticky='w')
        
        Label(self, text = "Source : ",font=('arial', 15, "bold")).grid(row=1,column=0,sticky='w')
        sourceList=['bhubaneswar','mumbai','bangalore','delhi','london']
        self.cb1=ttk.Combobox(self,values=sourceList)
        self.cb1.current(0)
        self.cb1.grid(row=1,column=1,sticky='w')
        
        Label(self, text = "Destination : ",font=('arial', 15, "bold")).grid(row=2,column=0,sticky='w')
        destList=['bhubaneswar','mumbai','bangalore','delhi','london']
        self.cb2=ttk.Combobox(self,values=destList)
        self.cb2.current(1)
        self.cb2.grid(row=2,column=1,sticky='w')
        
        self.cb3=None
        self.b3=None
        self.b4=None
        self.b5=None
        self.dojEntry=None
        
        self.b1=Button(self,text='search',bg='black',fg='white',width=10,command=self.showFlightList)
        self.b1.grid(row=3, column=0,sticky='w')
        self.b2=Button(self,text='back',bg='black',fg='white',width=10,command=lambda: master.SwitchFrame(UserWelcomePage))
        self.b2.grid(row=3,column=1,sticky='w')
        
    def showFlightList(self):
        '''shows the list of flights betwn a source and destination'''
        self.b1.destroy()
        self.b2.destroy()
        sdlist=(self.cb1.get(),self.cb2.get())
        try:
            mydb=sqlite3.connect("AirlineDatabase.db")
            cur=mydb.cursor()
            flist=[]
            temp=cur.execute("select * from flightInfo where source=? and destination=?",sdlist)
            while True:
                t=temp.fetchmany()
                if not t:
                    break
                flist.append(t[0][0]+","+t[0][1])
            mydb.commit()
            mydb.close()

            Label(self, text = "Choose Flight : ",font=('arial', 15, "bold")).grid(row=4,column=0,sticky='w')
            self.cb3=ttk.Combobox(self,values=flist)
            self.cb3.current(0)
            self.cb3.grid(row=4,column=1,sticky='w')
            
            Label(self, text = "Date of journey : ",font=('arial', 15, "bold")).grid(row=5,column=0,sticky="w")
            self.dojEntry = Entry(self, font=(5))
            self.dojEntry.grid(row=5, column=1,sticky="w")
        
            Button(self,text='confirm',bg='black',fg='white',width=10,command=self.confirmBooking)\
            .grid(row=6, column=0,sticky='w')

            Button(self,text='back',bg='black',fg='white',width=10,command=lambda: self.master.SwitchFrame(UserWelcomePage))\
            .grid(row=6, column=1,sticky='w')
        except:
            messagebox.showerror("Error","something went wrong")
        

    def confirmBooking(self):
        ''''Confirm the flight booking'''
        msg=messagebox.askquestion("Confirm Booking","Confirm Booking ?")
        if msg=="yes":
            try:
                mydb=sqlite3.connect("AirlineDatabase.db")
                cur=mydb.cursor()
                info=((self.cb3.get().split(','))[0],self.master.session,self.dojEntry.get())
                mydb.execute('INSERT INTO bookingInfo (flightId,userId,doj) VALUES(?,?,?)',info)
                mydb.commit()
                mydb.close()
                messagebox.showinfo("Booking Confirmed","Booking Confirmed")
                self.master.SwitchFrame(UserWelcomePage)
            except:
                messagebox.showerror("Error","something went wrong")
        else:
            self.master.SwitchFrame(BookFlight)

#----------------------------SEARCH BOOKINGS------------------------------------------------------
class  SearchBooking(Frame):
    def __init__(self, master):
        Frame.__init__(self, master)
        master.title("Search Booking")
        
        Label(self, text = "***** SEARCH BOOKINGS *****",font=('Helvetica', 18, "bold"),justify="center",\
              bg='white',fg='blue').grid(row=0,columnspan=2)
        try:
            mydb=sqlite3.connect("AirlineDatabase.db")
            cur=mydb.cursor()

            un=(self.master.session,)
            temp=cur.execute("select * from bookingInfo join flightInfo USING(flightId) where userId=?",un)
            Label(self, text ="BOOK ID",font=('arial', 10, "bold")).grid(row=3,column=0,sticky="w")
            Label(self, text ="FLIGHT ID",font=('arial', 10, "bold")).grid(row=3,column=1,sticky="w")
            Label(self, text ="SOURCE",font=('arial', 10, "bold")).grid(row=3,column=2,sticky="w")
            Label(self, text ="DESTINATION",font=('arial', 10, "bold")).grid(row=3,column=3,sticky="w")
            i=4
            while True:
                t=temp.fetchmany()
                if not t:
                    break
                Label(self, text = t[0][0],font=('arial', 10, "bold")).grid(row=i,column=0,sticky="w")
                Label(self, text = t[0][1],font=('arial', 10, "bold")).grid(row=i,column=1,sticky="w")
                Label(self, text = t[0][5],font=('arial', 10, "bold")).grid(row=i,column=2,sticky="w")
                Label(self, text = t[0][6],font=('arial', 10, "bold")).grid(row=i,column=3,sticky="w")
                i+=1

            mydb.commit()
            mydb.close()
            Button(self,text='back',bg='black',fg='white',width=10,\
                           command=lambda: master.SwitchFrame(UserWelcomePage)).grid(row=i,column=1,sticky='w')
        except:
            messagebox.showerror("Error","something went wrong")

#-------------------------DELETE BOOKING PAGE----------------------------------------------------------------------
class  DeleteBooking(Frame):
    def __init__(self, master):
        Frame.__init__(self, master)
        master.title("Delete Booking")
        
        Label(self, text = "***** DELETE BOOKINGS *****",font=('Helvetica', 18, "bold"),bg='white',fg='blue').grid(row=0,columnspan=2,sticky='w')        
        Label(self, text = "BookId : ",font=('arial', 15, "bold")).grid(row=1,column=0,sticky='w')
        self.BookIdEntry = Entry(self, font=(3),bg="powder blue")
        self.BookIdEntry.grid(row=1, column=1,sticky="w")
        
        self.b1=Button(self,text='delete',bg='black',fg='white',width=10,command=self.deleteBooking)
        self.b1.grid(row=2, column=0,sticky='w')
        
        self.b2=Button(self,text='back',bg='black',fg='white',width=10,\
                       command=lambda: master.SwitchFrame(UserWelcomePage))
        self.b2.grid(row=2, column=1,sticky='w')
        
    def deleteBooking(self):
        '''delete bookings'''
        msg=messagebox.askquestion("Confirm Deleting","Confirm Delete ?")
        if msg=="yes":
            try:
                mydb=sqlite3.connect("AirlineDatabase.db")
                cur=mydb.cursor()

                un=(self.master.session,self.BookIdEntry.get())
                cur.execute("DELETE from bookingInfo where userId=? and bookId=?",un)
                mydb.commit()
                mydb.close()
            except:
                messagebox.showerror("Error","something went wrong")
            else:
                messagebox.showinfo("Delete Successfull","Booking detail deleted....")
                self.master.SwitchFrame(UserWelcomePage)
        else:
            self.master.SwitchFrame(UserWelcomePage)

#-------------------------UPDATE USER DETAILS--------------------------------------------------------------------
class UpdateDetail(Frame):
    def __init__(self, master):
        Frame.__init__(self, master)
        master.title("Update User Details")
        
        Label(self, text = "***** Update Your details *****",font=('Helvetica', 18, "bold"),\
        justify="center",bg='white',fg='blue').grid(row=0,columnspan=2)
        
        Label(self, text = "Choose Option : ",font=('arial', 15, "bold"))\
        .grid(row=1,column=0)
        
        fieldList=['userName','userEmail','userMobile','useraddress']
        self.cb=ttk.Combobox(self,values=fieldList)
        self.cb.current(2)
        self.cb.grid(row=1,column=1)
        
        Label(self, text = "Enter Value : ",font=('arial', 15, "bold"))\
        .grid(row=2,column=0)

        self.updateValue = Entry(self, font=(2),bg='powder blue')
        self.updateValue.grid(row=2, column=1,sticky="w")
        
        Button(self,text='update',bg='black',fg='white',width=10,command=self.updateDetail)\
        .grid(row=3, column=0,sticky='w')
        
        Button(self,text='back',bg='black',fg='white',width=10,\
                       command=lambda: master.SwitchFrame(UserWelcomePage)).grid(row=3, column=1,sticky='w')

    def updateDetail(self):
        '''update user details'''
        msg=messagebox.askquestion("Confirm Update","Confirm Update ?")
        if msg=="yes":
            try:
                mydb=sqlite3.connect("AirlineDatabase.db")
                cur=mydb.cursor()
                if self.cb.get()=='userName':
                    ud=(self.updateValue.get(),self.master.session)
                    cur.execute("UPDATE userInfo SET userName = ? where userId=?",ud)
                elif self.cb.get()=='userMobile':
                    ud=(self.updateValue.get(),self.master.session)
                    cur.execute("UPDATE userInfo SET userMobile = ? where userId=?",ud)
                elif self.cb.get()=='userEmail':
                    ud=(self.updateValue.get(),self.master.session)
                    cur.execute("UPDATE userInfo SET userEmail = ? where userId=?",ud)
                elif self.cb.get()=='userAddress':
                    ud=(self.updateValue.get(),self.master.session)
                    cur.execute("UPDATE userInfo SET userAddress = ? where userId=?",ud)
                mydb.commit()
                mydb.close()
            except:
                messagebox.showerror("Error","something went wrong")
            else:
                messagebox.showinfo("Updation Successfull","User detail updated....")
                self.master.SwitchFrame(UserWelcomePage)
        else:
            self.master.SwitchFrame(UserWelcomePage)

        
#-------------------------AIRLINE ENQUIRY PAGE-------------------------------------------------------------------       
class  SearchAirline(Frame):
    def __init__(self, master):
        Frame.__init__(self, master)
        master.title("Airline")

        Label(self, text = "***** AIRLINE *****",font=('Helvetica', 18, "bold"),justify="center",bg='white',fg='blue')\
        .grid(row=0,columnspan=2)
        airlinesList=['indigo airlines','spice jet','luftansa airlines','vistara airlines','air india']
        self.cb=ttk.Combobox(self,values=airlinesList)
        self.cb.current(2)
        self.cb.grid(row=1,columnspan=4)
        
        self.b1=Button(self,text='search',bg='black',fg='white',width=10,command=self.showList)
        self.b1.grid(row=2, column=0,sticky='w')
        Button(self,text='back',bg='black',fg='white',width=10,command=lambda: master.SwitchFrame(UserWelcomePage)).\
        grid(row=2,column=1,sticky='w')
        
    def showList(self):
        '''shows the flights of a airline'''
        try:
            mydb=sqlite3.connect("AirlineDatabase.db")
            cur=mydb.cursor()
            an=(self.cb.get(),)
            temp=cur.execute("select * from flightInfo join airlineInfo USING(airlineName) where airlineName=?",an)
            i=5
            Label(self, text ="FLIGHT NO",font=('arial', 10, "bold")).grid(row=3,column=0,sticky="w")
            Label(self, text ="source",font=('arial', 10, "bold")).grid(row=3,column=1,sticky="w")
            Label(self, text ="Destin",font=('arial', 10, "bold")).grid(row=3,column=2,sticky="w")
            Label(self, text ="Type",font=('arial', 10, "bold")).grid(row=3,column=3,sticky="w")
            while True:
                t=temp.fetchmany()
                if not t:
                    break
                Label(self, text = t[0][0],font=('arial', 10, "bold")).grid(row=i,column=0,sticky="w")
                Label(self, text = t[0][2],font=('arial', 10, "bold")).grid(row=i,column=1,sticky="w")
                Label(self, text = t[0][3],font=('arial', 10, "bold")).grid(row=i,column=2,sticky="w")
                Label(self, text = t[0][4],font=('arial', 10, "bold")).grid(row=i,column=3,sticky="w")
                i+=1

            mydb.commit()
            mydb.close()
            self.b1.destroy()
        except:
            messagebox.showerror("Error","something went wrong")

#------------------------------Show User Information-----------------------------
class  UserDetail(Frame):
    def __init__(self, master):
        Frame.__init__(self, master)
        master.title("User Details")

        Label(self, text = "***** USER DETAIL *****",font=('Helvetica', 15, "bold"),bg='white',fg='blue').grid(row=0,columnspan=2)
        try:
            mydb=sqlite3.connect("AirlineDatabase.db")
            cur=mydb.cursor()
            uId=(self.master.session,)
            temp=cur.execute("select * from userInfo where userId=?",uId)

            listField=["ID","NAME","MOBILE","EMAIL","ADDRESS"]
            t=temp.fetchmany()
            for i in range(5):
                Label(self, text =listField[i],font=('arial', 10, "bold")).grid(row=3,column=i,sticky="w")
                Label(self, text = t[0][i],font=('arial', 10, "bold")).grid(row=4,column=i,sticky="w")
            mydb.commit()
            mydb.close()
            Button(self,text='back',bg='black',fg='white',width=10,\
                           command=lambda: master.SwitchFrame(UserWelcomePage)).grid(row=5,columnspan=2)
        except:
            messagebox.showerror("Error","something went wrong")
    
h=SetUp()
h.mainloop()    
```
